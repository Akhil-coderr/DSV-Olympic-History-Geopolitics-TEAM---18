```python
import kagglehub
```

The dataset uses a non-UTF-8 encoding. Kagglehub internally assumes UTF-8 and does not expose encoding parameters, so the file must be loaded using Pandas with the correct encoding,so from kaggle we only downloaded the filepath

```python
dataset_path = kagglehub.dataset_download(
    "heesoo37/120-years-of-olympic-history-athletes-and-results"
)

dataset_path
```

The dataset was saved using a different character encoding. Pandas assumes UTF-8 by default, which caused a decoding error. Specifying the correct encoding resolved the issue.

```python
athlete_file_path = dataset_path + "/athlete_events.csv"
noc_file_path = dataset_path +"/noc_regions.csv"
df_athlete = pd.read_csv(athlete_file_path, encoding="latin1")
df_noc = pd.read_csv(noc_file_path,encoding="latin1")
```

---



To add a second dataset to a main DataFrame,we can use merge when both datasets share a common key. merge is preferred in relational datasets because it preserves data integrity.

When working with multiple datasets, the `merge()` function is used when both datasets share a common column (key).  

In our dataset both dataframe contains common key i.e 'NOC'

This key helps combine related information from different sources into a single DataFrame.


---

```python
df=pd.merge(df_athlete, df_noc, on="NOC", how="left")
```
